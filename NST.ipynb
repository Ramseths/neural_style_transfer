# Neural Style Transfer
# Importar los paquetes necesarios

```
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt 
import numpy as np
import cv2
```

# Cargar el modelo pre-entrenado de TF Hub

```
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

# Preprocesamiento de Imágenes

```
def cargar_imagen(img_path):
  img = tf.io.read_file(img_path)
  img = tf.image.decode_image(img, channels=3)
  img = tf.image.convert_image_dtype(img, tf.float32)
  img = img[tf.newaxis, :]
  return img
```

# Cargar las imágenes

```
#Nuestra imagen a transformar
content_image = cargar_imagen('/content/drive/MyDrive/Computer_Image/profile.jpg')
#El estilo que deseamos
style_image = cargar_imagen('/content/drive/MyDrive/Computer_Image/vangogh.jpeg')
```

```
content_image.shape #Obtener la dimensión de la img
```

```
plt.imshow(np.squeeze(style_image))
#Mostrar el estilo
plt.show()
```

```
stylized_image = model(tf.constant(content_image), tf.constant(style_image))[0] #Entrenamiento
```

# Resultado

```
plt.imshow(np.squeeze(stylized_image))
plt.show()
```
